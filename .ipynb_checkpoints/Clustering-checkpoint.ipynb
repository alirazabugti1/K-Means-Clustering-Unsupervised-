```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data=pd.read_csv('income.csv')
```

```python
data.head()
```

```python
plt.scatter(data['Age'],data['Income($)'])
```

```python
km=KMeans(n_clusters=3)
```

```python
km.fit(data[['Age','Income($)']])
```

```python
predicted=km.predict(data[['Age','Income($)']])
```

```python
data['Clusters']=predicted
```

```python
data.head()
```

```python
data1=data[data['Clusters']==0]
data2=data[data['Clusters']==1]
data3=data[data['Clusters']==2]

plt.scatter(data1.Age,data1['Income($)'],color='green')
plt.scatter(data2.Age,data2['Income($)'],color='red')
plt.scatter(data3.Age,data3['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(data[['Income($)']])
data['Income($)']=scaler.transform(data[['Income($)']])


scaler.fit(data[['Age']])
data['Age']=scaler.transform(data[['Age']])
```

```python
data
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(data[['Age','Income($)']])
data['Clusters']=y_predicted
```

```python
data
```

```python
km.cluster_centers_
```

```python
data1=data[data['Clusters']==0]
data2=data[data['Clusters']==1]
data3=data[data['Clusters']==2]

plt.scatter(data1.Age,data1['Income($)'],color='green')
plt.scatter(data2.Age,data2['Income($)'],color='red')
plt.scatter(data3.Age,data3['Income($)'],color='black')

#for centroids

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```
