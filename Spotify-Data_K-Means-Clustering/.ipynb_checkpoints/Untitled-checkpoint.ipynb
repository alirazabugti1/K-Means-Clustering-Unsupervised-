```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('Spotify.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['track_name']=data['track_name'].fillna(data['track_name'].mode()[0])
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cmap='viridis')
```

```python
km=KMeans(n_clusters=3)
predicted=km.fit_predict(data[['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']])
```

```python
data['Clusters']=predicted
```

```python
data.head(5)
```

```python
# Apply PCA to reduce the dimensions to 2 for better visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(data[['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']])

# Add PCA results to the DataFrame
data['PCA1'] = pca_components[:, 0]
data['PCA2'] = pca_components[:, 1]

# Plotting the clusters in 2D space (PCA1 vs PCA2)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Clusters', data=data, palette='viridis', s=100, alpha=0.7)

# Add labels and title
plt.title('Clusters of Songs Based on Audio Features (PCA Reduced)', fontsize=14)
plt.xlabel('PCA Component 1', fontsize=12)
plt.ylabel('PCA Component 2', fontsize=12)
plt.legend(title='Clusters', loc='best')

# Show the plot
plt.show()
```
